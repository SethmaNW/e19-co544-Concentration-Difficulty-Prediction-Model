```python
import pandas as pd  # frames
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv('../Preprocessing_pipeline/final_data_sets/train_set.csv')

# Display the DataFrame
print(df)
```

```python
df.dtypes
```

SVM Model

Train with initial data

Taking x_train

```python
x_train = df.drop(["Difficulty_level"], axis=1)
x_train.head()
```

Take y_train

```python
y_train = df["Difficulty_level"]
y_train.value_counts()
```

```python
from sklearn import svm

classifier = svm.SVC(kernel="linear", gamma="auto", C=2)

classifier.fit(x_train, y_train)

```

Taking test data

```python
# Read the CSV file into a DataFrame
test_df = pd.read_csv("../Preprocessing_pipeline/final_data_sets/test_set.csv")

test_df.head()
```

```python
x_test = test_df.drop(["Difficulty_level"], axis=1)
```

```python
y_test = test_df["Difficulty_level"]
```

```python
y_predict = classifier.predict(x_test)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, classification_report

accuracy = accuracy_score(y_test, y_predict)
print(classification_report(y_test, y_predict))
print(accuracy)
```
